```python
# Set the path to the folder containing the files
folder_path = '/content/drive/My Drive/bdm_final_project/olist_data_for_bdm/'

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the order items dataset
order_items = pd.read_csv(folder_path + 'olist_order_items_dataset.csv')

# Calculate the revenue per order item (price + freight value)
order_items['revenue'] = order_items['price'] + order_items['freight_value']

# Aggregate the total revenue and the number of orders per seller
seller_performance = order_items.groupby('seller_id').agg({'revenue':'sum', 'order_id':'nunique'}).reset_index()

# Rename columns for clarity
seller_performance.rename(columns={'order_id':'number_of_orders'}, inplace=True)

# Plotting the seller performance
# Here we'll create two subplots: one for the number of orders and one for the total revenue

plt.figure(figsize=(14, 6))



# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(seller_performance['number_of_orders'], seller_performance['revenue'], alpha=0.5)
plt.title('Seller Performance: Number of Orders vs. Total Revenue')
plt.xlabel('Number of Orders')
plt.ylabel('Total Revenue')
plt.xscale('log')  # Optional: Apply log scale for better visibility if data is skewed
plt.yscale('log')  # Optional: Apply log scale for better visibility if data is skewed
plt.grid(True)
plt.show()
```
