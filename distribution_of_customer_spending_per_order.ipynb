```python
# Set the path to the folder containing your files
folder_path = '/content/drive/My Drive/bdm_final_project/olist_data_for_bdm/'

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the order items dataset
order_items = pd.read_csv(folder_path + 'olist_order_items_dataset.csv')

# Calculate the total spending per order by summing the price and freight value
order_items['total_spending'] = order_items['price'] + order_items['freight_value']

# Now, we'll create a histogram of total spending per order
plt.figure(figsize=(12, 6))
sns.histplot(order_items['total_spending'], bins=50, kde=False)
plt.title('Distribution of Customer Spending per Order')
plt.xlabel('Total Spending')
plt.ylabel('Number of Orders')
plt.xlim(0, order_items['total_spending'].quantile(0.95))  # Limiting to 95th percentile for better visualization
plt.show()
```
